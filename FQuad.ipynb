```python
import json
```

```python
thedata = open('train.json', encoding='utf8').read()
data = json.loads(thedata)
```

```python
data = data['data']
```

```python
len(data)
```

```python
the_questions = []
the_answers = []
```

```python
for i in range(len(data)):
    try:
        paragraph = data[i]['paragraphs'][i]
        question_count = str(paragraph).count("'question'")
        for q in range(question_count):
            try:
                questions = paragraph['qas'][q]['question']
                answer = paragraph['qas'][q]['answers']
                the_questions.append(questions)
                the_answers.append(answer)
            except:
                pass
    except:
        pass
```

```python
final_questions = []
final_answers = []
for i in range(len(the_questions)):
    try:
        ques = the_questions[i]
        answ = the_answers[i][0]['text']
        print("Question:",ques)
        print("Answer:",answ)
        final_questions.append(ques)
        final_answers.append(answ)
        print('')
        print('------------------------------------')
        print('')
    except:
        pass
```

```python
len(final_answers), len(final_questions)
```

### Preparing Dataframe for data analysis

```python
import pandas as pd

data = {'Questions': final_questions, 'Answers': final_answers}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
from transformers import XLMTokenizer, XLMForQuestionAnsweringSimple
```

```python
import torch
```

```python
tokenizer = XLMTokenizer.from_pretrained('xlm-mlm-en-2048')
```

```python
model = XLMForQuestionAnsweringSimple.from_pretrained('xlm-mlm-en-2048')
```

```python
inputs = tokenizer(list(df['Questions']), list(df['Answers']), return_tensors = 'pt', padding=True, truncation=True)
inputs
```

```python
start_positions = torch.tensor([1])
```

```python
end_positions = torch.tensor([3])
```

```python
outputs = model(**inputs, start_positions=start_positions, end_positions=end_positions)
```

```python
loss = outputs.loss
```

```python
start_scores = outputs.start_logits
```

```python
end_scores = outputs.end_logits
```
